## Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)

    Goal_diff_count = GoalsFor - GoalsAgainst

```python
def Goal_diff_count(data):
    
    goal_difference = []
    for index, value in zip(data.index,data['Team']):
        
        goal_difference.append((value,data.GoalsFor[index] - data.GoalsAgainst[index]))
#         print(f'Team {value} has an overall goal difference of {data.GoalsFor[index] - data.GoalsAgainst[index]}')
    return goal_difference
```

```python
Goal_diff_count(data)
```

```python
# maximum goal difference
```

```python
# use absolute in lambda because some negative goal difference can be bigger than that of a pozitive goal difference team!
sorted(Goal_diff_count(data),key=lambda x : abs(x[1]),reverse=True)[0]
```

```python
# minimum goal difference 
```

```python
sorted(Goal_diff_count(data),key=lambda x : abs(x[1]),reverse=True)[-1]
```

##  Create a new column with name “Winning Percent” and append it to the data set (5 points)

Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage

```python
def winning(data):
    
    winning_percent = []
    for index in data.index:
        winning_percent.append((data.GamesWon[index] /data.GamesPlayed[index])*100)
#         print(f'Team {value} has an overall goal difference of {data.GoalsFor[index] - data.GoalsAgainst[index]}')
    return winning_percent
```

```python
data['WinningPercent'] = winning(data)
```

```python
data['WinningPercent'] = data['WinningPercent'].fillna(0)
```
